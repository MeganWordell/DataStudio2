```python
import pandas as pd

# Read the first CSV file into a pandas DataFrame
df1 = pd.read_csv('neighborhoods.csv')

# Read the second CSV file into a pandas DataFrame
df2 = pd.read_csv('incomes.csv')

# Merge the two DataFrames on the ID column
merged_df = pd.merge(df1, df2, on='name')

# Write the merged DataFrame to a new CSV file
merged_df.to_csv('merged.csv', index=False)
```

```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('merged.csv')

# Group the DataFrame by the 'name' column
grouped_df = df.groupby('NTAType')

# Concatenate the groups into a single DataFrame
merged_df = pd.concat([grouped_df.get_group(name) for name in grouped_df.groups])

# Write the resulting DataFrame to a new CSV file
merged_df.to_csv('grouped_data.csv', index=False)
```

```python
import geopandas as gpd

# Read the GeoJSON file into a geopandas GeoDataFrame
gdf = gpd.read_file('query-2.json')

# Filter the GeoDataFrame to only include certain features
filtered_gdf = gdf[gdf['BoroName'] == 'Manhattan']

# Write the filtered GeoDataFrame to a new GeoJSON file
filtered_gdf.to_file('filtered_file.json', driver='GeoJSON')
```
